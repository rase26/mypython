## -- jieba斷詞
- 匯入自定義字典
- 匯入停用字字典

```python
#使用jieba斷詞
jieba.set_dictionary('dict.txt.big') #匯入繁體字典
jieba.load_userdict('userdict.txt')  #匯入自定義ˋ字典

# 匯入中文的停用字字典 變成 stopWords list
stopWords=[]
with open('Chinese Stop Words_t.txt', 'r', encoding='utf8') as f: 
    for data in f.readlines():
        data = data.strip()
        stopWords.append(data)

#由於一些符號 直接在txt檔輸入 無法被辨識出來 所以直接append到stopWords上 
stopWords.append(' ') 
stopWords.append('\r\n')
stopWords.append('.')
stopWords.append('\r')
stopWords.append('\t')
stopWords.append('')
stopWords.append('●')
stopWords.append('or')
stopWords.append('of')
stopWords.append('. ')
stopWords.append('•')
stopWords.append('【')
stopWords.append('】')
stopWords.append('‧')
stopWords.append('／')
stopWords.append('…')
stopWords.append(' & ')
stopWords.append('★')
stopWords.append('.  ')
stopWords.append('  ')
stopWords.append('－')
stopWords.append('「')
stopWords.append('www')
stopWords.append('　')
stopWords.append('nan')
```

### --建立新增英文詞的函數:雙字組成的詞 因為在自定義字典中無法匯入

```python
def add_eng(w): ##新增英文詞
    jieba.re_han_default = re.compile("([\u4E00-\u9FD5a-zA-Z0-9+#&\._% ]+)", re.U)
    jieba.add_word(w)
```

```python
#新增英文詞
add_eng('machine learning')
add_eng('deep learning')
add_eng('data mining')
add_eng('scikit learn')
add_eng('data analysis')
add_eng('power bi')
add_eng('big data')
jieba.add_word('tensorflow')
```

### --建立文字雲所需的詞彙清單的函數

```python
#建立文字雲所需list
def makewordlist(df):
    #取df 的'工作內容' '條件要求' '工具要求'
    wordlist=pd.Series()
    wordlist=wordlist.append(df['工作內容']).append(df['條件要求']).append(df['工具要求'])
    kw=[]
    #斷詞
    for t in wordlist:
        tt=str(t)
        ks=jieba.cut(tt.lower())
        kw.extend(ks)

    #過濾停用字
    rkw= list(filter(lambda a: a not in stopWords and a != '\n', kw))
    return rkw
```

### --建立詞彙清單詞頻的函數

```python
#計算詞頻
def wordfreq(data):
    ck=pd.Series() #變成 Series 可以去算詞頻
    for i in data:
        ck=ck.append(pd.Series(i))

    sc=ck.value_counts() #算詞頻 
    return sc
    
```

### --建立相關職缺 詞表及其詞頻

```python
#資料科學相關職缺
#建立詞表
rkw_data = makewordlist(df_data)
#計算詞頻
sc_data=wordfreq(rkw_data)
```

```python
#機器學習相關職缺
#建立詞表
rkw_ml = makewordlist(df_ml)
#計算詞頻
sc_ml=wordfreq(rkw_ml)
```

### --建立產生文字雲的函數

```python
def toolwordcloud(data,mask,filename):
    import matplotlib.pyplot as plt
    from wordcloud import WordCloud
    import random
    from palettable.colorbrewer.qualitative import Dark2_8 #字詞的顏色色盤 可以自己上網找喜歡的做更改
    
    #定義字詞顏色色盤
    def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
        return tuple(Dark2_8.colors[random.randint(0,7)]) #有八個顏色去做random
    
    word_could_dict=Counter(data) #data 要經過計數
    bg_pic = np.array(Image.open(mask)) #要用這張圖當輪廓
    
    #建立文字雲
    wordcloud = WordCloud(width = 1000, height = 500,font_path="msjhbd.ttc",max_words=400,mask=bg_pic,background_color="white").generate_from_frequencies(word_could_dict)
    #換字的顏色
    wordcloud.recolor(color_func=color_func, random_state=2) 
    #存圖
    wordcloud.to_file(filename)
    #顯示圖
    plt.figure(figsize=(15,8))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
    plt.close()
```
