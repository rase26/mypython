# 1.  建立資料庫

### -- 建立網址參數的函數:用於搜尋不同關鍵字及頁數

```python
##網頁參數的建立:用於搜尋頁
def searchjob(key,r,p): #key:關鍵字 r: 1=全職 2=兼職 只搜尋職務有關鍵字的 p:頁數
    global params
    params={"ro":r,"kwop":1,"keyword":key,"order":1,"asc":0,"scstrict":0,"scneg":1,"page":p,"mode":"s"}
```

### -- 建立爬取資料網頁的函數:抓取職缺的相關資訊

```python
##資料的爬取
def jobinfo(res): #爬工作的資料
    #將變數變得外面也可以用
    global job,company,salary,exp,indust,relist,toolist,jcontent,date,jlink

    p=BeautifulSoup(res.text,'html.parser')

    job = pd.Series() #職位 
    company = pd.Series() #公司 
    indust =pd.Series() #產業別 
    toolist=pd.Series() #擅長工具
    relist=pd.Series() #條件要求
    jcontent=pd.Series() #工作內容
    jlink =pd.Series() #職缺網址
    salary = pd.Series() #起薪
    exp= pd.Series() #工作經歷 
    date=pd.Series() #postdate

    print('開始抓了~')

    for i in p.select("article.b-block--top-bord.job-list-item.b-clearfix.js-job-item"): 
        #職位 公司 網址資訊都在這一塊裡面，所以用迴圈一個一個抓
        clink='https:'+ i.select("h2.b-tit a")[0].get('href') #職缺網址
        print(clink)
        if "relevance" in clink: #不相關的職位不要放入
            job = job.append(pd.Series([i['data-job-name']])).reset_index(drop=True) #職位
            company = company.append(pd.Series([i['data-cust-name']])).reset_index(drop=True) #公司
            indust=indust.append(pd.Series([i['data-indcat-desc']])).reset_index(drop=True) #產業別

            jc0=i.select('p.job-list-item__info.b-clearfix.b-content')
            if len(jc0)!=0:
                jcontent=jcontent.append(pd.Series([jc0[0].text])).reset_index(drop=True) #工作內容
            else:
                jcontent=jcontent.append(pd.Series(["NA"])).reset_index(drop=True)
            jlink=jlink.append(pd.Series([clink])).reset_index(drop=True) #網址
            p2=BeautifulSoup(requests.get(str(clink)).text,"html.parser")#職缺網址內容

            #條件要求
            re0=p2.select('section.info')
            if len(re0)!=0:
                re1=re0[1].select('div.content dl dd')[-1].text
                relist=relist.append(pd.Series([re1])).reset_index(drop=True)
            
            #讀不到的話 再讀一次 通常可以讀的到資料 再讀不到只好存入NA
            elif len(re0) ==0:
                p2=BeautifulSoup(requests.get(str(clink)).text,"html.parser")
                re0=p2.select('section.info')
                if len(re0)!=0:
                    re1=re0[1].select('div.content dl dd')[-1].text#條件要求
                    relist=relist.append(pd.Series([re1])).reset_index(drop=True)
                else:
                    relist=relist.append(pd.Series(['NA'])).reset_index(drop=True)

            tlist= p2.select('dd.tool a') #擅長工具List 還要用迴圈抓 .text
            tool=''
            for t in tlist:
                tool+=' '+t.text
            toolist=toolist.append(pd.Series([tool])).reset_index(drop=True)

            #薪資 只看起薪
            s=i.select("span.b-tag--default")[0].text[2:]
            if s=='面議': #目前規定面議薪水要4萬以上，故以4萬計算
                s0=40000
            else: #起薪
                s=s.split('~')
                s0=int(s[0].replace(',',''))
            salary = salary.append(pd.Series([s0])).reset_index(drop=True)

            #所需工作經驗
            explist=i.select('ul.b-list-inline.b-clearfix.job-list-intro.b-content li')
            #因為爬出來理面每三個的第二個才是我們需要的,所以用迴圈抓取
            n=1
            for e in explist:
                e=e.text
                if e != '':
                    if n%3==0:
                        n=1
                    elif n%2== 0:
                        exp=exp.append(pd.Series([e])).reset_index(drop=True)
                        n+=1
                    else:
                        n+=1
            #工作刊登時間
            dl=i.select('span.b-tit__date')[0].text.strip().split("/")
            d="2018-"+dl[0]+"-"+dl[1] #轉成資料庫的時間格式
            date=date.append(pd.Series([d])).reset_index(drop=True)

            print('正在抓取資料中')
    print('資料抓取完成')
```

### -- 建立將資料匯入資料庫的函數

```python
#匯入mysql資料庫
def importomysql(k,job,company,salary,exp,indust,relist,toolist,jcontent,date,jlink):
    #連接資料庫
    dlink = pymysql.connect(
                host ="localhost",
                user="root",
                passwd="",
                db="job" #資料庫名稱
                )
    #執行資料庫的變數
    cmd = dlink.cursor()

    ##直接加入 jobinfo 中的迴圈
    n=0
    
    #將資料list 裡面資料一個一個取出
    for j in job:
        c=company[n]
        s=str(salary[n])
        e=exp[n]
        i=indust[n]
        r=relist[n]
        t=toolist[n]
        jc=jcontent[n]
        d=date[n]
        l=jlink[n]

        if cmd.execute("select * from `104` where `jobtitle`= %s and `company`=%s",(j,c)) == 0: #查詢資料庫 若資料庫沒有則匯入
            #匯入資料庫的資料
            pm= {
                "keyword":k,
                "jobtitle":j,
                "date":d,
                "company":c,
                "salary":s,
                "experiencetime":e,
                "industry":i,
                "requirement":r,
                "requiretool":t,
                "jobcontent":jc,
                "joblink":l
                }
            
            #執行匯入資料庫 新增列
            cmd.execute("INSERT INTO `104`(`keyword`,`jobtitle`,`date`,`company`,`salary`,`experiencetime`,`industry`,`requirement`,`requiretool`,`jobcontent`,`link`) VALUES(%(keyword)s,%(jobtitle)s,%(date)s,%(company)s,%(salary)s,%(experiencetime)s,%(industry)s,%(requirement)s,%(requiretool)s,%(jobcontent)s,%(joblink)s)",pm)
            dlink.commit #執行
            print('寫入mysql中', n)
            
        elif cmd.execute("select `id`,`keyword` from `104` where `jobtitle`= %s and `company`=%s",(j,c)) !=0: #已有此筆資料，但關鍵字不同，新增關鍵字
            k0=cmd.fetchall()[0] #取出資料已有的關鍵字資訊
            Id=k0[0]     #那筆資料的id
            if k not in k0[1] : #如果沒有新的關鍵字 則新增關鍵字
                k1=k0[1]+' '+k
                cmd.execute("update `104` set `keyword`=%s where `jobtitle`= %s and `company`=%s",(k1,j,c))
                dlink.commit()
                print('mysql中已有此筆資料,在id=%s,新增關鍵字:%s,%s'%(Id,k,n))
            else: #已經有關鍵字了 就不用新增了
                print('mysql中已有此筆資料及關鍵字:%s,在id=%s,%s'%(k,Id,n))
        elif cmd.execute("select `id`,`keyword` from `104` where `jobtitle`= %s and `company`=%s",(j,c)) != 0:
            k0=cmd.fetchall()[0]
            Id=k0[0]
            print('mysql中已有此筆資料,在id=%s,%s'%(Id,n))
        else:
            print('有問題!!')
        n+=1

    dlink.close()
    print('成功寫入mysql')
```

## - 從104銀行爬取資料 並匯入資料庫

```python
#關鍵字清單
k_data=['大數據','資料分析','資料科學','巨量分析','巨量']  #資料科學 數據分析相關
k_ml=['機器學習','深度學習','人工智慧','AI','AI 智能']     #機器學習 人工智慧相關
```

```python
#爬取 資料科學 數據分析相關 職缺資訊 匯入資料庫

for k in k_data:#用迴圈一次一次抓關鍵字清單 的關鍵字
    for p in range(1,10): #只抓十頁
        searchjob(k,1,p) #只抓全職的職缺
        res = requests.get("https://www.104.com.tw/jobs/search/",params)
        print('正在抓取關鍵字: %s, 第 %s 頁'%(k,p))
        jobinfo(res)
        importomysql(k,job,company,salary,exp,indust,relist,toolist,jcontent,date,jlink)
```

```python
#### 爬取 機器學習 人工智慧相關 職缺資訊 匯入資料庫

for k in k_ml:#用迴圈一次一次抓關鍵字清單 的關鍵字
    for p in range(1,10): #只抓十頁
        searchjob(k,1,p) #只抓全職的職缺
        res = requests.get("https://www.104.com.tw/jobs/search/",params)
        print('正在抓取關鍵字: %s, 第 %s 頁'%(k,p))
        jobinfo(res)
        importomysql(k,job,company,salary,exp,indust,relist,toolist,jcontent,date,jlink)
```

```python
#從資料庫內匯入資料變成dataframe

dlink = pymysql.connect(
                host ="localhost",
                user="root",
                passwd="",
                db="job" #資料庫名稱
                )
q = dlink.cursor()

##建立 資料科學 數據分析相關 職缺資訊 dataframe
df_data=pd.DataFrame({
    '職缺':[],'公司':[],'薪資':[],'經歷年資':[],'工作內容':[],'條件要求':[],'工具要求':[],'產業別':[]
    })
k1='%'+k_data[0]+'%'
k2='%'+k_data[1]+'%'
k3='%'+k_data[2]+'%'
k4='%'+k_data[3]+'%'
k5='%'+k_data[4]+'%' 

l=q.execute("select * from `104` where `keyword` like %s or `keyword` like %s or `keyword` like %s or `keyword` like %s or `keyword` like %s",(k1,k2,k3,k4,k5))
y=q.fetchall()
i=0
for d in y:
    df_data.ix[i,'職缺']= y[i][2]
    df_data.ix[i,'公司']= y[i][4]
    df_data.ix[i,'薪資']= y[i][5]
    df_data.ix[i,'經歷年資']=y[i][6]
    df_data.ix[i,'工作內容']= y[i][10]
    df_data.ix[i,'條件要求']= y[i][8]
    df_data.ix[i,'工具要求']= y[i][9]
    df_data.ix[i,'產業別']= y[i][7]
    i+=1


##建立 機器學習 人工智慧相關 職缺資訊 dataframe

df_ml=pd.DataFrame({
    '職缺':[],'公司':[],'薪資':[],'經歷年資':[],'工作內容':[],'條件要求':[],'工具要求':[],'產業別':[]
    })
k1='%'+k_ml[0]+'%'
k2='%'+k_ml[1]+'%'
k3='%'+k_ml[2]+'%'
k4='%'+k_ml[3]+'%'
k5='%'+k_ml[4]+'%' 

l=q.execute("select * from `104` where `keyword` like %s or `keyword` like %s or `keyword` like %s or `keyword` like %s or `keyword` like %s",(k1,k2,k3,k4,k5))
y=q.fetchall()
i=0
for d in y:
    df_ml.ix[i,'職缺']= y[i][2]
    df_ml.ix[i,'公司']= y[i][4]
    df_ml.ix[i,'薪資']= y[i][5]
    df_ml.ix[i,'經歷年資']=y[i][6]
    df_ml.ix[i,'工作內容']= y[i][10]
    df_ml.ix[i,'條件要求']= y[i][8]
    df_ml.ix[i,'工具要求']= y[i][9]
    df_ml.ix[i,'產業別']= y[i][7]
    i+=1
    

dlink.close()
```

### 技能的職缺數長條圖

```python
##畫出數量前幾名的長條圖
def counts(df,column,title,pngname,r):
    #計算提到的次數 存成字典
    b=[]
    count={}
    for tt in df[column] :
        t=tt.split(' ')
        for i in t:
            if i not in b:
                count.update({i:1})
                b.append(i)
            else:
                count[i]+=1
    if '' in count:
        del count[''] #刪掉空白值
    count=sorted(count.items(), key=lambda item: item[1], reverse=True) #排序後 變成list
    i=0
    names=[]
    plt.figure(figsize=(16,6)) #畫長條圖 只取前十名
    for i in range(r):
        plt.bar(i,count[i][1],tick_label=count[i][0])
        names+=[count[i][0]]

    plt.xticks(range(0,r),names)
    plt.title(title)
    plt.savefig(pngname)
    plt.show()
    return count
```
